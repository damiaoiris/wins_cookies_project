```python
import pandas as pd
import os
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
```

### Prepare data for analysis 

**site_visit_url:** websited the bot has visited

**visit_id:** it says if the visit to the website is the same 

**crawl:** This data base only has non - disinformation websites (pilot_1_US_base) and disinformation websites (pilot_1_US_misinfo)

**is_third_party:** cookie set by the domain visited (first-party) set by another domain = third_party (third-party)

**disconnect_tracker_category:** if the third-party is associated with some type of content/function (Advertisement, Social,Fingerprinting...) that information is in this column. If none NaN

**companies:** Big company behind the third-party

```python
path = os.getcwd() 
data_path = path + "/companies_cookie_data_copy.csv"
df = pd.read_csv(data_path, sep=";")
df
```

```python
from sklearn.preprocessing import LabelEncoder

new_df = df.loc[df['is_third_party'] == 'third_party']
new_df["url_id"] = LabelEncoder().fit_transform(new_df["site_visit_url"])
new_df["tp_id"] = LabelEncoder().fit_transform(new_df["ps_plus_1"])
new_df[['site_visit_url', 'ps_plus_1', 'url_id','tp_id']]

new_df['tp_id'] = new_df['tp_id'].apply(lambda x : str(x))
new_df['url_id'] = new_df['url_id'].apply(lambda x : str(x))
new_df['texts'] = new_df.groupby(['url_id'])['tp_id'].transform(lambda x: ' '.join(x))
new_df = new_df[['url_id', 'texts']].drop_duplicates(subset='url_id')
new_df
```

```python
titles = new_df['url_id'].tolist()
corpus = new_df['texts'].tolist()
```

```python
file = open('titles.txt','w')
for item in titles:
	file.write(item+"\n")
file.close()
```

```python
file = open('corpus.txt','w')
for item in corpus:
	file.write(item+"\n")
file.close()
```

```python
titles = [str(title) for title in list(np.unique(new_df['url_id']))]
corpus = []
for title in list(np.unique(new_df['url_id'])):
    doc = str()
    temp_df = new_df.loc[new_df['url_id'] == title]
    for i, word in enumerate(list(temp_df['tp_id'])):
        w = str(word)
        if i < len(list(temp_df['tp_id']))-1:
            doc = doc + w + " "
        else:
            doc = doc + w
    corpus.append(doc)
```
