```python
#imports

import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# data 
path = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
data_path = path + "/data/cookie_data_clean.csv"
df = pd.read_csv(data_path, sep=";")
```

### Relevant columns in the dataset

**site_visit_url:** websites the bot has visited

**crawl:** This database only has non - disinformation websites (pilot_1_US_base) and disinformation websites (pilot_1_US_misinfo)

**is_third_party:** cookie set by the domain visited (first-party) set by another domain = third_party (third-party)

**ps_plus_1:** domain of the third-party website

**disconnect_tracker_category:** if the third-party is associated with some type of content/function (Advertisement, Social,Fingerprinting...) that information is in this column. If none NaN

**companies:** Big company behind the third-party

## Data Cleaning

```python
df = df.loc[df['is_third_party'] == 'third_party']

# assign a unique ID to the websites and the third-party companies
df["url_id"] = LabelEncoder().fit_transform(df["site_visit_url"])
df["tp_id"] = LabelEncoder().fit_transform(df["ps_plus_1"])
```

```python
# store the mappings into dictionaries 
mapping_sites = dict(zip(df.url_id, df.site_visit_url))     # url_id -> url
mapping_crawl = dict(zip(df.url_id, df.crawl))              # url_id -> crawl (misinfo or not)
mapping_crawlV2 = dict(zip(df.site_visit_url, df.crawl))
mapping_tp = dict(zip(df.tp_id, df.ps_plus_1))              # third-party id -> domain
mapping_tp_reverse = dict(zip(df.ps_plus_1, df.tp_id))      # domain -> third-party id
mapping_companies = dict(zip(df.tp_id, df.companies))       # third-party id -> main company
mapping_ctype_num = dict(zip(df.tp_id, df.disconnect_tracker_category))
```

```python
# maybe not necessay, but convert the IDs into strings
df['url_id'] = df['url_id'].apply(lambda x : str(x))
df['tp_id'] = df['tp_id'].apply(lambda x : str(x))
```

```python
# create a new column called 'texts' where we flatten all the third-party companies acting on each website
df['texts'] = df.groupby(['site_visit_url'])['ps_plus_1'].transform(lambda x: ' '.join(x))

# drop duplicated columns
df = df.drop_duplicates(subset='site_visit_url')

# edgelist
df[['site_visit_url', 'texts']].head(10)
```

```python
# nodes on the left
titles = df['site_visit_url'].tolist()
titles = [h.split()[0] for h in titles]

# nodes on the right
corpus = df['texts'].tolist()
texts = [h.split() for h in corpus]
```

## TopSBM: Topic Modeling with Stochastic Block Models

```python
from sbmtm import sbmtm
import graph_tool.all as gt
```

```python
# fit data
model = sbmtm()

## word-document network from the corpus
model.make_graph(texts,documents=titles)

## fit the model
gt.seed_rng(32) ## seed for graph-tool's random number generator --> same results
model.fit()
```

```python
# assign colors to nodes based on their attributes

colors = {
    'pilot_1_US_base': '#50A2A7',
    'pilot_1_US_misinfo': '#9B2915',
    'Google': '#E9B44C',
    'Yandex': '#5762D5',
    'AddThis': '#AA9ABA'
}

big_companies = ['Google', 'Yandex', 'AddThis', 'Facebook', 'WPP', 'PubMatic',
       'Rocket Fuel', 'Microsoft', 'SMI2', 'comScore', 'Yahoo!']

vertex_colors = [colors[c] for c in mapping_crawl.values()] + [colors[c] if c in colors.keys() else 'white' for c in mapping_companies.values()]
```

```python
# assign node attributes to the graph

# vertex name
vertex_text = model.g.vp.name.get_2d_array([0])[0]
model.state.g.vp['vertex_text'] = model.state.g.new_vp('string', vals = vertex_text)

# vertex color and size
model.state.g.vp['vertex_color'] = model.state.g.new_vp('string', vals = vertex_colors)
model.state.g.vp['vertex_size'] = model.state.g.new_vp('int', val = 5)
```

```python
# plot

model.plot(nedges=10000, 
           vertex_size = model.g.vp.vertex_size,
           vertex_fill_color = model.g.vp.vertex_color,
           vertex_color = 'white',          
           bg_color ='white',
           output_size = (900,200))


plt.show()
```

# Bar plots for communities on cookies

```python
# calculate frequencies for the cookies bar plot

ct_frequencies = {}
for level in range(5):
    ct_frequencies[level] = {}
    for c, comm in enumerate(model.topics(l=level, n=10000).keys()):
        ad = 0
        analy = 0
        cont = 0
        discont = 0
        FG = 0
        FI = 0
        soc = 0
        none = 0
        deno = len(model.topics(l=level, n=10000)[c])
        for n, node in enumerate(model.topics(l=level, n=10000)[c]):
            key = int(mapping_tp_reverse[node[0]])
            if mapping_ctype_num[key] == 'Advertising':
                ad += 1
            elif mapping_ctype_num[key] == 'Analytics':
                analy += 1
            elif mapping_ctype_num[key] == 'Content':
                cont += 1
            elif mapping_ctype_num[key] == 'Disconnect':
                discont += 1
            elif mapping_ctype_num[key] == 'FingerprintingGeneral':
                FG += 1
            elif mapping_ctype_num[key] == 'FingerprintingInvasive':
                FI += 1
            elif mapping_ctype_num[key] == 'Social':
                soc += 1
            else:
                none += 1


        ct_frequencies[level][c] = [ad/deno, analy/deno, cont/deno, discont/deno, FG/deno, FI/deno, soc/deno, none/deno]
```

```python
ct_values = list(ct_frequencies[3].values())
ct_values
rearrange = {}
key_list = ['Advertising', 'Analytics', 'Content', 'Disconnect', 'FingerprintingGeneral', 'FingerprintingInvasive', 'Social', 'None']
for i in range(8):
  key = key_list[i]
  rearrange[key] = [x[i] for x in ct_values]

rearrange2 = rearrange
rearrange2.pop('None')

#plot without 'none' column
colors_res = {
    'Advertising': '#50A2A7',
    'Analytics': '#9B2915',
    'Content': '#E9B44C',
    'Disconnect': '#5762D5',
    'FingerprintingGeneral': '#AA9ABA',
    'FingerprintingInvasive': '#556545',
    'Social': '#9d581c'
}

community = ['Community 1', 'Community 2', 'Community 3', 'Community 4', 'Community 5', 'Community 6']
x = np.arange(len(community))
width = 0.1  # the width of the bars
multiplier = 0

print(x)

fig, ax = plt.subplots(layout='constrained', figsize = (10,5))
for cookie_type, measurement in rearrange2.items():
    offset = width * multiplier
    rects = ax.bar(x + offset , measurement, width, label=cookie_type, color = colors_res[cookie_type])
    #ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percent of Community')
ax.set_title('Type of Cookie')
ax.set_xticks(x+(width), community)
ax.legend(ncols=1)

plt.savefig('Type_noNone.png',dpi = 600)

plt.show()
```

# Bar plots for communities on websites

```python
df=df.reset_index(drop=True)
frequencies = {}
for level in range(5):
    frequencies[level] = {}
    for c, comm in enumerate(model.clusters(l=level, n=10000).keys()):
        fake = 0
        base = 0
        deno = len(model.clusters(l=level, n=10000)[c])
        for n, node in enumerate(model.clusters(l=level, n=10000)[c]):
            key = node[0]
            if mapping_crawlV2[key] == 'pilot_1_US_base':
                base += 1
            else:
                fake += 1
        frequencies[level][c] = [deno, (fake/deno,base/deno), (fake, base)]
```

```python
def prop_variance(n):
    p = 0.5593
    q = 1-p
    variance = p*q/n
    return variance
level = 4
community = ["Community " + str(k+1) for k in frequencies[level].keys()]
community = ['Population']+community
fake = [round(frequencies[level][k][1][0],2) for k in frequencies[level].keys()]
fake=[0.56]+fake
base = [round(frequencies[level][k][1][1],2) for k in frequencies[level].keys()]
base=[0.44]+base
fake_v = [prop_variance(frequencies[level][k][0]) for k in frequencies[level].keys()]
base_v = fake_v
species = tuple(community)
penguin_means = {
    'Misinfo': tuple(fake),
    'Non-Misinfo': tuple(base),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0
fig, ax = plt.subplots(layout='constrained')
colors = ['#9B2915','#50A2A7']
for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute,color = colors[multiplier])
    ax.bar_label(rects, padding=3, color = colors[multiplier], fontsize=12)
    multiplier += 1
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Proportion', fontsize=15)
ax.set_title('Distribution of misinformation websites by' +'\n'+ 'community', fontsize=16, pad=20)
ax.set_xticks(x + 0.125, species)
ax.tick_params(labelsize=15)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.legend(loc='upper left', ncols=3, fontsize=13)
ax.set_ylim(0, 1)
#plt.savefig('Websites.pdf')
plt.show()
```
